```python
import matplotlib.pyplot as plt
def plotme(data):
  dictionary = data
  xAxis = [key for key, value in dictionary.items()]
  yAxis = [value for key, value in dictionary.items()]
  return (xAxis,yAxis)
```

```python
try:
  response = natural_language_understanding.analyze(
    text=data,
    features=Features(emotion=EmotionOptions(targets=['apples','oranges']))).get_result()
  print(json.dumps(response, indent=2))
except:
  print ("Error processing the data for emotional analysis")
```

```python
try:
  X,Y= plotme((response['emotion']['document']['emotion']))
  #plt.grid(True)
  fig = plt.figure()
  plt.bar(X,Y, color='maroon')
  plt.xlabel('emotion')
  plt.ylabel('value')
  plt.show()
except:
  print ( "Error ploting graph for Emotional analysis")
```

```python
response = natural_language_understanding.analyze(
    text=data,
    features=Features(concepts=ConceptsOptions(limit=3))).get_result()

print(json.dumps(response, indent=2))
```

```python
import pandas as pd 

df = pd.DataFrame(response['concepts'])
df.plot.bar(x='text',y='relevance')
```

```python
response = natural_language_understanding.analyze(
    text=data,
    features=Features(categories=CategoriesOptions(limit=3))).get_result()

print(json.dumps(response, indent=2))
```

```python
df = pd.DataFrame(response['categories'])
df.plot(x='label',y='score')
```
