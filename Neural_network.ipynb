Creating sigmoul 

```python
def sigmoul (x, deriv=False):
  if(deriv==True):
    return(x*(1-x))
  return(1/(1+np.exp(-x)))
```

Back prapogration
#seed 

```python
np.random.seed(2)
```

#synapses


```python
syn0 = 2*np.random.random((5,50)) - 1
syn1 = 2*np.random.random((50,1)) - 1 
```

#Training

```python
for j in range(2000):
  #layer
  l0 = array
  l1 = sigmoul(np.dot(l0,syn0))
  l2 = sigmoul(np.dot(l1,syn1))
  #backprogation
  l2_error = target - l2
  if (j % 100) == 0:
    print ('Error ' + str(np.mean(np.abs(l2_error))))
  #calculate delta
  l2_delta = l2_error*sigmoul(l2, deriv=True)
  l1_error = l2_delta.dot(syn1.T)
  l1_delta = l1_error * sigmoul(l1, deriv=True)
  #update weight
  syn1 += l1.T.dot(l2_delta)
  syn0 += l0.T.dot(l1_delta)
```
